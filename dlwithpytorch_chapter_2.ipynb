The ```resent``` variable can be called as function, it just taking as input single or multiple images and providing an equal number of fully connected 1000 ImageNet classes. Before we do that, we need to preporcess the input images so they may get the right size and the value within the same neumerical range. To do so, ```trochvision``` module provides us ```transforms```, which allow us to quickly define the basic preprocessing function:

````python
from torchvision import transforms
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean = [0.485, 0.456, 0.406],
        std = [0.229, 0.224, 0.225]
    )
])
````

We can now grab a picture of our favorite teacher, preprocess it, and then see what ResNet thinks of it. We can start by loading an image from the local filesystem using ```Pillow``` , an image-manipulation module for Python:

At this point, we need to determine the index corresponding to the maximum score in the out tensor we obtained previously. We can do that using the max function in PyTorch, which outputs the maximum value in a tensor as well as the indices where that maximum value occurred:

````python
_, index = torch.max(out, 1)
````

````python
percentange = torch.nn.functional.softmax(out, dim=1)[0] * 100
labels[index[0]], percentange[index[0]].item()
````
